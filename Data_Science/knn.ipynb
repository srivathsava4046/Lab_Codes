```python
import math
import matplotlib.pyplot as plt

# Define the dataset
dataset = [(2, 3, 'Class A'), (3, 5, 'Class A'), (5, 3, 'Class A'), (6, 2, 'Class A'),
           (2, 6, 'Class B'), (4, 8, 'Class B'), (6, 7, 'Class B'), (7, 9, 'Class B')]

# Implement the KNN classifier
def knn_classify(x, y, k):
    distances = []
    for data_point in dataset:
        x_i, y_i, label = data_point
        distance = math.sqrt((x - x_i)**2 + (y - y_i)**2)
        distances.append((distance, label))

    # Sort the distances and select the top K neighbors
    distances.sort()
    k_nearest_neighbors = distances[:k]

    # Count the occurrences of each class among the K neighbors
    class_counts = {}
    for _, label in k_nearest_neighbors:
        class_counts[label] = class_counts.get(label, 0) + 1

    # Determine the majority class
    predicted_class = max(class_counts, key=class_counts.get)
    return predicted_class

# Test the KNN classifier with a specific data point
new_data_point = (4,6)  # Change this to test with different data points
k = 2  # Change this to adjust the number of neighbors to consider

# Create a scatter plot to visualize the data points and KNN classification
plt.figure(figsize=(8, 6))

# Plot data points and their classes
for x_i, y_i, label in dataset:
    color = 'r' if label == 'Class A' else 'g'
    plt.scatter(x_i, y_i, c=color, marker='o', label=label)

# Highlight the new data point
plt.scatter(new_data_point[0], new_data_point[1], c='b', marker='x', label='New Data Point')

# Determine the predicted class for the new data point
predicted_class = knn_classify(new_data_point[0], new_data_point[1], k)

print(f"Predicted class for data point {new_data_point}: {predicted_class}")

plt.title(f'KNN Classification (k={k}) - Predicted Class: {predicted_class}')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
#plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a KNN classifier with a specified number of neighbors (e.g., K=3)
k = 3
knn_classifier = KNeighborsClassifier(n_neighbors=k)

# Fit the classifier on the training data
knn_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = knn_classifier.predict(X_test)

# Calculate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
