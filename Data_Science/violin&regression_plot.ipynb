```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
data.head()
```

```python
slength = data['SepalLengthCm']
swidth = data['SepalWidthCm']
plenght = data['PetalLengthCm']
pwidth = data['PetalWidthCm']
species = data['Species']
```

```python
setosa_data = data.loc[data['Species'] == 'Iris-setosa']
vercicolor_data = data.loc[data['Species'] == 'Iris-versicolor']
virginica_data = data.loc[data['Species'] == 'Iris-virginica']
```

```python
# Create a violin plot
sns.violinplot(x=species, y=slength, data=data)

# Add labels and title
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.title("Violin Plot of Sepal Length by Species")

# Show the plot
plt.show()
```

**Regression plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 4, 5, 4, 6, 8, 7, 9, 10]

# Create a regression plot
sns.regplot(x=x, y=y, scatter_kws={"s": 100}, ci=None)

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regression Plot")

# Show the plot
plt.show()
```
