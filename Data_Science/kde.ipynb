```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
data.head(100)
```

```python
slength = data['SepalLengthCm']
swidth = data['SepalWidthCm']
plenght = data['PetalLengthCm']
pwidth = data['PetalWidthCm']
species = data['Species']
```

```python
setosa_data = data.loc[data['Species'] == 'Iris-setosa']
vercicolor_data = data.loc[data['Species'] == 'Iris-versicolor']
virginica_data = data.loc[data['Species'] == 'Iris-virginica']
```

```python
sns.kdeplot(setosa_data['SepalLengthCm'],color='b', shade=True, label='Setosa')
sns.kdeplot(virginica_data['SepalLengthCm'],color='r', shade=True, label='Virginica')
sns.kdeplot(vercicolor_data['SepalLengthCm'],color='g', shade=True, label='vercicolor')
plt.legend()
plt.xlabel('Sepal Length')
plt.ylabel('Length in cm')
```
